```python
#Rename columns for consistency + easier feature engineering/modelling


df = df.rename(columns = {
    'User_ID': 'user_id',
    'Product_ID': 'product_id',
    'Gender': 'gender',
    'Age': 'age_group',
    'Occupation': 'occupation',
    'City_Category': 'city_cat',
    'Stay_In_Current_City_Years': 'years_in_city',
    'Marital_Status': 'marital_status',
    'Product_Category_1': 'product_cat1',
    'Product_Category_2': 'product_cat2',
    'Product_Category_3': 'product_cat3',
    'Purchase': 'purchase_amt'
    
})
df
```

```python
#Convert low-cardinality columns to 'category' dtype,
#Enables efficient groupby/pivot operations


cat_cols = [
    'gender',
    'age_group',
    'city_cat',
    'years_in_city',
    'marital_status',
    'occupation'
]

df[cat_cols] = df[cat_cols].astype('category')
```

```python
# Fill product_cat2/3 NaNs with 'Unknown' 
# Reason: Dropping ~58% of 550k rows loses critical customer-product signals
# Alternative (mode imputation) would bias toward popular categories

df['product_cat2'] = df['product_cat2'].fillna('Unknown')
df['product_cat3'] = df['product_cat3'].fillna('Unknown')
df
```

```python
# Check remaining nulls in product category columns after imputation
df[['product_cat1', 'product_cat2', 'product_cat3']].isna().sum()
```

```python
# Convert filled product_cat2/3 to 'category' dtype
df[['product_cat2', 'product_cat3']] = df[['product_cat2', 'product_cat3']].astype('category')
df.info() # Verify memory savings
```
